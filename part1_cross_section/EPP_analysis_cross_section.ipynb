```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 14
```

# Analysis

## MET 

```python
df_MET = pd.read_json("scores/analysis/MET_1.json")
df_MET = df_MET[1:]
df_MET
```

```python
# Plots
plt.errorbar(x=df_MET.cut_MET, y=df_MET.score_W, yerr=df_MET.score_W_err, fmt="o", capsize=2.5)
plt.ylim(0,15)
plt.title("Missing transverse energy")
plt.xlabel("MET [GeV]")
plt.ylabel("Score")
plt.savefig("plots/score_plots/Plot_MET.pdf")
```

## MuonPT

```python
df_PT = pd.read_json("scores/analysis/MuonPT_0.json")
df_PT = df_PT[1:]
df_PT
```

## NJets
