```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.size'] = 14
```

```python
df = pd.DataFrame()
df1 = pd.read_json(r"data/Results_asymmetry_0_0.4.json")
df2 = pd.read_json(r"data/Results_asymmetry_0.4_0.8.json")
df3 = pd.read_json(r"data/Results_asymmetry_0.8_1.5.json")
df4 = pd.read_json(r"data/Results_asymmetry_1.5_1.8.json")
df5 = pd.read_json(r"data/Results_asymmetry_1.8_2.1.json")
```

```python
df = pd.concat([df, df1])
df = pd.concat([df, df2])
df = pd.concat([df, df3])
df = pd.concat([df, df4])
df = pd.concat([df, df5])
df = df.reset_index()
df = df.drop(columns=["index"])
```

```python
asymmetry = df.asymmetry.apply(pd.Series, index=['asym', 'asym_err'])
asymmetry["Eta"] = ["[0.0, 0.4]", "[0.4, 0.8]","[0.8, 1.5]","[1.5, 1.8]","[1.8, 2.1]"]
asymmetry
```

```python
plt.errorbar(x = asymmetry.Eta, y = asymmetry.asym, yerr = asymmetry.asym_err, fmt="o", capsize=2.5)
plt.title("$\mathcal{A} (\eta)$")
plt.xlabel("Range of the pseudorapidity $\eta$")
plt.ylabel("W charge asymmetry $\mathcal{A}$")
plt.savefig("plots/A_eta.pdf")
```
